error of the mean model$  = {y_i} - {y_{mean}}$ 


If you calculate this error for each value of y and then calculate the sum of the square of each error, you will get a quantity that is proportional to the variance in y. It is known as the **Total Sum of Square** (TSS).

Do you know the formula for variance calculation?
its effectively , $Var(y)=TSS/N$ given $ TSS = \sum_{i=0}^n ({y_i}-\bar y) ^ 2$

Because **TSS/N** is the actual variance in y, the TSS is proportional to the total variance in your data.

if you fit a Linear Regression Model to the above data by using the OLS technique, i.e. by minimizing the sum of squares of residual errors (RSS), the worst that you can do is the Mean Model.   
But the sum of squares of residual errors of the Mean Model is simply TSS, i.e. for the Mean Model, RSS = TSS.

The model fitting procedure of these nonlinear models is **not based on progressively minimizing the sum of squares of residual errors (RSS)** and therefore the **optimally fitted model could have a residual sum of squares that is greater than total sum of squares**. That means, RÂ² for such models can be a negative quantity. As such, RÂ² is not a useful goodness-of-fit measure for most nonlinear models.

A notable exception is regression models that are fitted using the Nonlinear Least Squares (NLS) estimation technique. The NLS estimator seeks to minimizes the sum of squares of residual errors thereby making RÂ² applicable to NLS regression models.
